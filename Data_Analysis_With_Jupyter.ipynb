### Vẽ đồ thị hàm sin,cos,tan trong khoảng -2𝜋 -> 2𝜋

```python
import math
pi_arr = np.linspace(-2*math.pi,2*math.pi)
```

```python
plt.plot(pi_arr,np.sin(pi_arr))
plt.title('Đồ thị hàm sin trong khoảng -2π -> 2π')
```

```python
plt.plot(pi_arr,np.cos(pi_arr))
plt.title('Đồ thị hàm cos trong khoảng -2π -> 2π')
```

```python
plt.plot(pi_arr,np.tan(pi_arr))
plt.title('Đồ thị hàm tan trong khoảng -2π -> 2π')
```

```python
plt.plot(pi_arr,np.sin(pi_arr))
plt.plot(pi_arr,np.cos(pi_arr))
plt.plot(pi_arr,np.tan(pi_arr))
plt.title('Đồ thị hàm sin,cos,tan trong khoảng -2π -> 2π')
plt.legend(['y = sin(-2π -> 2π)','y = cos(-2π -> 2π)','y = tan(-2π -> 2π)'])
```

### Tạo 1000 dữ liệu kiểu int ngẫu nhiêu gắn với 1000 ngày từ 09/02/2010.

### Tính cumsum() và vẽ đồ thị.


```python
ts = pd.Series(data=np.random.randint(1000, size=1000), index=pd.date_range(start='09/02/2010', periods=1000))
ts
```

```python
ts = ts.cumsum()
ts.head()
```

```python
ts.plot()
plt.title('Đồ thị hàm cumsum() của ts')
```

### Đọc file CSV dientich.csv (trong thư mục này) bằng pandas: pd.read_csv()
### Tìm ra 10 tỉnh có diện tích lớn nhất Việt Nam năm 2015.
### Vẽ đồ thị cột (bar) diện tích của 10 tỉnh đó.
#### (dữ liệu lấy từ tổng cục thống kê Việt Nam: https://www.gso.gov.vn/default.aspx?tabid=714 chọn "Diện tích, dân số và mật độ dân số phân theo địa phương").


```python
df = pd.read_csv('dientich.csv',sep=';',header=1,index_col=0,skiprows=[3]).dropna()
df
```

```python
max_acreage = df['2015'].sort_values(ascending=True).tail(10)
max_acreage
```

```python
max_acreage.plot.bar()
plt.title('Đồ thị 10 tỉnh có diện tích lớn nhất Việt Nam năm 2015')
plt.legend(['Diện tích'])
```
