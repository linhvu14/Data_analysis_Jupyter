```python
plt.plot(pi_arr,np.cos(pi_arr))
plt.title('Äá»“ thá»‹ hÃ m cos trong khoáº£ng -2Ï€ -> 2Ï€')
```

```python
plt.plot(pi_arr,np.tan(pi_arr))
plt.title('Äá»“ thá»‹ hÃ m tan trong khoáº£ng -2Ï€ -> 2Ï€')
```

```python
plt.plot(pi_arr,np.sin(pi_arr))
plt.plot(pi_arr,np.cos(pi_arr))
plt.plot(pi_arr,np.tan(pi_arr))
plt.title('Äá»“ thá»‹ hÃ m sin,cos,tan trong khoáº£ng -2Ï€ -> 2Ï€')
plt.legend(['y = sin(-2Ï€ -> 2Ï€)','y = cos(-2Ï€ -> 2Ï€)','y = tan(-2Ï€ -> 2Ï€)'])
```
